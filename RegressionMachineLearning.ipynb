## Regression Machine Learning
It is a supervised learning technique because the Dataset used contains some features and Outcomes(Label).
Regression focuses on predicting labels that are numerical in nature. e.g., Sales prediction

```python
# SALES PREDICTION BASED ON ADVERTISING METHODS
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# READING THE SALES DATASET
sales_data = pd.read_csv('https://msi.martial.co.ke/data/Advertising.csv')
sales_data.head(5)
```

```python
# DATA ANALYSIS
sales_data.info()
```

```python
# Checking Empties
sales_data.isnull().sum()
```

```python
# Statistical Analysis
sales_data.describe()
```

```python
# Correlation Analysis
# TV Ads -> 78 percent of sales
# Radio -> 57 percent
# Newspapers -> 22 percent
# Feature elimination
sales_data.corr()
```

```python
# MACHINE LEARNING
# STEP 1: DIVIDE THE DATASET INTO FEATURES(X) AND LABELS(Y)
# Convert the Sales Dataframe unto Numpy Arrays so that we can extract its values
array = sales_data.values
array
```

```python
# X = Features(TV, Radio, Newspapers)
# 200 Records and 3 Columns(TV, Radio, Newspapers)
X = array[:, 0:3]
X.shape
```

```python
# Y = Label(sales)
Y = array[:, 3]
Y.shape
```

```python
# STEP 2: DIVIDE THE X AND Y INTO TRAINING(70%) AND TESTING(30%)
#X -> X train(70%) and X_test(30%)
#Y -> Y_train(70%) and Y_test(30%)

from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# STEP 3: IMPORTING THE SCIENTIFIC MACHINE LEARNING ALGORITHMS
# a) LINEAR REGRESSION
from sklearn.linear_model import LinearRegression

# b) DECISION TREE REGRESSION
from sklearn.tree import DecisionTreeRegressor

# c) K-NEAREST NEIGHBORS ALGORITHMS
from sklearn.neighbors import KNeighborsRegressor

# d) SUPPORT VECTOR REGRESSION
from sklearn.svm import SVR

# e) RANDOM FOREST
from sklearn.ensemble import RandomForestRegressor

# f) GRADIENT BOOSTING
from sklearn.ensemble import GradientBoostingRegressor

# g) ADABOOST
from sklearn.ensemble import AdaBoostRegressor
```

```python
# STEP 4: CROSS VALIDATION
# Hyperparameter Tuning
# The process that is used to determine the best learning algorithm for a particular problem
# This process validates and picks the right algorithm that will produce a higher accuracy
models = []
models.append(('D Trees', DecisionTreeRegressor()))
models.append(('KNR', KNeighborsRegressor()))
models.append(('LR', LinearRegression()))
models.append(('Support Machines', SVR()))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Gradient Boosting', GradientBoostingRegressor()))
models.append(('Support Machines', SVR()))

# Import Cross Validation and KFOLD
from sklearn.model_selection import cross_val_score, KFold
# Create a for loop so that each model is tested in turn
for name, model in models:
          # Here we do a 10 split K-FOLD
          kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

          # We get the results for each Fold, Train and Test using the Folds
          cv_results = cross_val_score(model, X_train, Y_train, cv = kfold,
          scoring='r2')

          # Get the average of all Folds for current model in the loop
          print(name, ' Results:= ', cv_results.mean())
```

```python
# STEP 5: MODEL TRAINING
# In this process the model learns from the training dataset using the best algorithm
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

```python
# STEP 6: MODEL EVALUATION
# This is the process used to Determine whether the model is accurate to predict the results
# In Regression(numbers) we use the Mean Absolute Error(mae)
# To evaluate a model, we will provide the testing features(X_test), the hide the testing outcomes(Y_test).
# The model will provide its own predictions(model_predictions) that will be compared against the correct predictions(Y_test)

model_predictions = model.predict(X_test)
print("The  model predicted: ",model_predictions)
```

```python
print("The correct predictions: ",Y_test)

# Therefore we wil import a scientific metric called r_2 score which is the ratio between the model predictions and the Y_test results
```

```python
# SCIENTIFIC MODEL EVALUATION
# REGRESSION: r2_score
# Ratio between the model predictions and Y_test
from sklearn.metrics import r2_score
print ("The accuracy of the model is: ", r2_score(Y_test, model_predictions)* 100 )
```
